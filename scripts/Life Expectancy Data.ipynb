```python
import pandas as pd
import matplotlib.pylab as plt
import os
import numpy as np
os.getcwd()
data = pd.read_csv('C:\\Users\\jacks\\Desktop\\Life Expectancy Data.csv')
data.head()
df = pd.DataFrame(data)
columns = list(df)
print(columns)
```

```python
data['Schooling'].isna().value_counts(normalize = True)
```

```python
school_vs_lifeexp = data[['Country', 'Life expectancy ', "Schooling"]]
print(school_vs_lifeexp)
#here I have created a table with three columns: country, life expect, and schooling
schooling_df = pd.DataFrame(school_vs_lifeexp)
#schooling_df is a DataFrame with only the columns country, lifeexpectancy, and schooling
print(schooling_df.shape)
print(schooling_df)
```

```python
data["Schooling"].isna().value_counts()
#There are 163 missing values in the schooling column
```

```python
print(data["Life expectancy "].isna().value_counts())
#there are 10 missing values in the "Life expectancy " column
data["Country"].isna().value_counts()
#no missing values in country column
```

```python
schooling_df.dropna(how = "any", inplace = True)
print(schooling_df)
print(schooling_df["Schooling"].isna().value_counts())
print(schooling_df["Life expectancy "].isna().value_counts())
#now we have created a data frame where we have removed all null values
```

```python
print(schooling_df.corr())
plt.scatter(schooling_df['Schooling'], schooling_df['Life expectancy '], color = "g")
model = np.poly1d(np.polyfit(schooling_df['Schooling'], schooling_df['Life expectancy '], 1))
polyline = np.linspace(0,22)
plt.plot(polyline, model(polyline))
print(model)
plt.title('Life Expectancy vs Schooling')
plt.xlabel('Schooling')
plt.ylabel("Life Expectancy")
plt.show()
```
